```python
import sys
import os
sys.path.insert(0, os.path.abspath('../'))

import numpy as np
import matplotlib.pyplot as plt

import astropy.units as u
import poliastro 

import CtllDes 
from CtllDes.core import ctll, satellite

from poliastro.bodies import Earth

sat = satellite.Sat.from_vectors([8000,0,0]*u.km,
                       [0,5,2.5]*u.km/u.s,
                      attractor=Earth)


```

```python
from CtllDes.targets.targets import Target
gs = Target(0,0)
```

```python
from CtllDes.requests.comm import default_comm_data
```

```python
default_df = default_comm_data(sat,gs,5,dt=5,J2=True,drag=False)
default_df
```

```python
times = default_df['time'].to_numpy(dtype=np.float64)/60 
elevations = default_df['elevation'].to_numpy(dtype=np.float64)*180/np.pi
```

```python
import matplotlib.pyplot as plt
%matplotlib qt5

fig, ax = plt.subplots(ncols=1, nrows=2, figsize=(10,10))

ax[0].grid()
ax[1].grid()

ax[0].plot(times,elevations,linewidth=1,color='k')
ax[0].set_title("Elevation vs. time for ground station located at (0°,0°)")
ax[0].set_xlabel("time [min]")
ax[0].set_ylabel("elevation [°]")


ax[1].plot(times,elevations,linewidth=1,color='orange')
ax[1].set_xlabel("time [min]")
ax[1].set_ylabel("elevation [°]")
```
